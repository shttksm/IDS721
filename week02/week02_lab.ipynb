## Terraform Fundamentals

**Terraform**
* Infrastructure as Code:
  * 記述には高レベルの構成構文
  * ブループリントをバージョン管理
  * インフラストラクチャを共有したり再利用可
* 実行プラン: 
  * 実行プランを生成する「プランニング」のステップがあり
* リソースグラフ: 
  * リソースのグラフが作成
  * インフラ内のリソース依存関係把握可
  * 依存関係ないリソースはパラに生成設定されて効率化
* 変更の自動化: 
  * 何がどの順番で変更されるかがわかる
  * 多くの人的ミスを回避

2. Terraform のインストールの確認

```console
terraform
```

3. 構成ファイル作成
* [公式ドキュメント](https://www.terraform.io/docs/language/index.html)
```
resource "google_compute_instance" "default" {
  project      = "<PROJECT_ID>"
  name         = "terraform"
  machine_type = "n1-standard-1"
  zone         = "us-central1-a"

  boot_disk {
    initialize_params {
      image = "debian-cloud/debian-9"
    }
  }

  network_interface {
    network = "default"
    access_config {
    }
  }
}
```
* リソースタイプは google_compute_instance
* google_compute_instance と入力すると、Google プロバイダが管理するリソースであることを Terraform が自動的に認識
